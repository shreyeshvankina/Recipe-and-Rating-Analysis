```python
# TODO
raw_recipes = pd.read_csv('RAW_recipes.csv')
interactions = pd.read_csv('interactions.csv')
merged = raw_recipes.merge(interactions, left_on='id', right_on='recipe_id', how='left')
merged['rating'] = merged['rating'].replace(0, np.nan)
avg_rating = merged.groupby('id')['rating'].mean()
recipes_with_avg = raw_recipes.copy()
recipes_with_avg['average_rating'] = recipes_with_avg['id'].map(avg_rating)
recipes_with_avg
```
